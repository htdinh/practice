```python
import matplotlib.pyplot as plt
import numpy as np
def log_likelihood(theta, theta_hats):
    """theta: true, theta_hat: estimated"""
    return [theta*np.log10(theta_hat)+(1-theta)*np.log10(1-theta_hat) for theta_hat in theta_hats]

STEP = 0.1
theta = 0.7
for theta in [0.4, 0.7]:
    theta_hats = list(np.arange(0, 1, STEP))
    logs = log_likelihood(theta, theta_hats)
    plt.plot(theta_hats, logs, label=r"All \theta")
    plt.plot(theta, log_likelihood(theta, [theta]), "o", label=r'\theta=\theta_hat'.format(theta))

plt.grid()
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0., fontsize=18)
plt.xlabel(r"theta_hat")
plt.ylabel("log likelihood")
plt.show()
```
